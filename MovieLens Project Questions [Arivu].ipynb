### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# grouping movie_df by movie id and adding all counts
movie_agg_df=rating_df.groupby("movie id").count() 


# grouping movie_df by movie id and adding all sums
movie_agg_df["rating_sum"]=rating_df.groupby("movie id").rating.sum()
movie_agg_df
```

```python
# Filtering only the movies with 100+ ratings and storing them in top_movie_agg dataframe
top_movie_agg=movie_agg_df[movie_agg_df["rating"]>=100]

# Adding avg rating as column to the same dataframe
top_movie_agg["avg_rating"]=top_movie_agg["rating_sum"]*1.00/top_movie_agg["rating"]*1.00

# sorting by avg rating and filtering only the top 25 and storing them in a separate dataframe top25_movie_agg
top25_movie_agg=top_movie_agg.sort_values(by="avg_rating",ascending = False).head(25)

```

```python
# Prepping the movie_df for joining and setting the movie id as index for easier join on movie id
movie_df_lookup=movie_df[["movie id","movie title"]]
movie_df_lookup.index = movie_df_lookup["movie id"]

# Joining the Top 25 movie avg rating with movie_df for the final report
Final_TOP25_Report=top25_movie_agg.join (movie_df_lookup,how="left")[["movie title", "avg_rating"]]

Final_TOP25_Report.reset_index()
```
